В рамках данного проекта проводится анализ вакансий "Аналитик данных" и "Специалист по Data Science" с сайта HH.ru.

**Цель проекта**

Основной целью проекта является выявление различий в предлагаемых вакансиях на позиции "Аналитик данных" и "Специалист по Data Science".

**Описание данных**

Для проведения анализа были использованы два датасета:
- `vacancies_da.xlsx` - содержит данные о вакансиях на позицию "Аналитик данных";
- `vacancies_ds.xlsx` - содержит данные о вакансиях на позицию "Специалист по Data Science".

Оба датасета содержат идентичный набор столбцов, в которых представлена следующая информация:
- id - уникальный идентификатор вакансии размещённой на сайте;
- name - название вакансии;
- published_at - дата и время публикации вакансии;
- alternate_url - ссылка на веб-страницу вакансии;
- type - статус вакансии на момент получения данных от API и передачи их в базу;
- employer - название компании-работодателя, разместившей вакансию;
- department - подразделение / отдел компании-работодателя;
- area - местоположение / регион компании-работодателя;
- experience - требования к опыту работы для соискателей вакансии (необходимый "грейд" - Junior, Middle, Senior и их подгрейды);
- key_skills - ключевые навыки для соискателей вакансии, полученные при анализе полного текста вакансии. Поле генерируется после получения информации от API;
- schedule - график работы; 
- employment - тип занятости;
- description - описание вакансии;
- description_lemmatized - лемматизированное описание вакансии, используемое для упрощения поиска аналогичных вакансий на сайте;
- salary_from - нижний порог заработной платы, предлагаемый компанией-работодателем (нижняя граница "зарплатной вилки");
- salary_to - верхний порог заработной платы, предлагаемый компанией-работодателем (верхняя граница "зарплатной вилки");
- salary_bin - уровень заработной платы;
- key_skills_from_key_skills_field - отформатированный перечень ключевых навыков для соискателей вакансии из столбца key_skills;
- hard_skills_from_description - "твёрдые" навыки, найденные при обработке полей с навыками. Поле генерируется после получения информации от API;
- soft_skills_from_description - "мягкие" навыки, найденные при обработке полей с навыками. Поле генерируется после получения информации от API.

```python
vacancies_da_ds[vacancies_da_ds.duplicated('id', keep=False)].sort_values(by='id')
```

Так как представленные вакансии относятся сразу к обоим направлениям, **удаление дубликатов в данном случае не требуется, поскольку информация и требования представленные в них могут анализироваться в отдельности**.

Поскольку идентичный номер id, дата и время публикации и ссылка на веб-страницу вакансии в рассмотренном случае не обязательно говорят о том, что вакансия является дубликатом, проверим наличие дубликатов сразу по всем столбцам, исключая `id`, `published_at` и `alternate_url`.

```python
vacancies_da_ds[vacancies_da_ds.duplicated(
    subset=[
    'name', 'employer', 'department', 'area', 'experience', 'key_skills', 'schedule', 'employment',
    'description', 'description_lemmatized', 'salary_from', 'salary_to', 'salary_bin',
    'key_skills_from_key_skills_field', 'hard_skills_from_description', 'soft_skills_from_description', 'vacancy_type'], keep=False
    )].sort_values(by=['name', 'employer', 'area', 'experience', 'published_at'])
```

Полученная выборка данных свидетельствует о наличии **637 строк с неявными дубликатами вакансий, которые различаются между собой только id, датой и временем и публикации и ссылкой на веб-страницу, что говорит о том, что их могли несколько раз переопубликовать**.

Удалим выявленные неявные дубликаты при помощи метода `drop_duplicates()`, которому передадим перечень столбцов в которых совпадают данные. При этом оставим наиболее актуальные вакансии, так как данные в них могли быть скорректированы по отношению к предыдущим публикациям вакансий.

```python
vacancies_da_ds = (
    vacancies_da_ds.drop_duplicates(
        subset=[
        'name', 'employer', 'department', 'area', 'experience', 'key_skills', 'schedule', 'employment',
        'description', 'description_lemmatized', 'salary_from', 'salary_to', 'salary_bin',
        'key_skills_from_key_skills_field', 'hard_skills_from_description', 'soft_skills_from_description', 'vacancy_type'],
        keep='last'
        )
)
```

```python
def get_unique_values(dataset, column):
    """Выводим и подсчитываем уникальные значения в столбце
    
    Передаем функции датасет и столбец который хотим просмотреть.
    Выводим уникальные значения в столбце и их количество при помощи метода value_counts.
    Подсчитываем количество уникальных значений в столбце при помощи метода nunique.
    Подсчитываем количество уникальных значений в столбце при удалении лишних пробелов
    в начале и в конце названия и при его приведении к нижнему регистру при помощи методов str.lower и str.strip.
    
    """
    display(dataset[column].value_counts())
    print('Количество уникальных значений в столбце:', dataset[column].nunique())
    print('Количество уникальных значений в столбце после форматирования:',
        dataset[column].str.lower().str.strip().nunique())
```

Проверим - есть ли вакансии, у которых столбец `key_skills` был заполнен, тогда как в столбце `key_skills_from_key_skills_field` значение отсутствовало. Для этого воспользуемся методами `query()`, `notna()` и `isna()`.

```python
# Выбираем строки из столбца key_skills, где нет пропущенных значений,
# и строки из столбца key_skills_from_key_skills_field, где есть пропущенные значения,
# а затем выбираем данные столбцы и подсчитываем их количество 
(vacancies_da_ds
    .query('key_skills.notna() & key_skills_from_key_skills_field.isna()')
    [['key_skills', 'key_skills_from_key_skills_field']]
).count()
```

Согласно полученным данным в датасете присутствует 1 015 строк, которые заполнены столбце `key_skills` и при этом не заполнены в столбце `key_skills_from_key_skills_field`. Как уже было упомянуто ранее, в столбце `key_skills` присутствуют пустые списки, что равносильно пропускам. Преобразуем их в пустые значения при помощи лямбда-функции и метода `replace()`.

```python
# Перезаписываем столбец key_skills, применяя к нему лямбда функцию, которая удалит все символы списков из строк
vacancies_da_ds['key_skills'] = (
    vacancies_da_ds['key_skills']
    .apply(lambda x: x
           .replace('"', '')
           .replace('[', '')
           .replace(']', '')
           .replace("'", ""))
)
     
# Проверяем успешность удаления представленных выше символов
vacancies_da_ds['key_skills'].value_counts()
```

Преобразования в столбце `key_skills` проведены успешно - теперь в нём присутствует 477 пустых строк, ранее являвшихся пустыми списками. Строки с данными значениями необходимо сопоставить с соответствующими строками столбце `key_skills_from_key_skills_field` и заменить их пропусками при помощи атрибута `loc[]`.

```python
# Сопоставляем пропущенные значения из столбца key_skills со значениями из столбца key_skills_from_key_skills_field
vacancies_da_ds.loc[
    (vacancies_da_ds['key_skills'] != '') & (vacancies_da_ds['key_skills_from_key_skills_field'].isna()),
    'key_skills_from_key_skills_field'
] = \
vacancies_da_ds.loc[
    (vacancies_da_ds['key_skills'] != '') & (vacancies_da_ds['key_skills_from_key_skills_field'].isna()),
    'key_skills'
]

# Подсчитываем количество пропусков в столбце key_skills_from_key_skills_field, оставшиеся после преобразований
vacancies_da_ds['key_skills_from_key_skills_field'].isna().sum()
```

Таким образом в столбце `key_skills_from_key_skills_field` осталось 477 пропущенных значений из 1 015.

При анализе информации, представленной в датасете, были зафиксированы случаи, когда при наличии значений в столбцах `salary_from` и `salary_to` в столбце `salary_bin` вместо указания уровня заработной платы установлено значение `ЗП не указано`. Исправим это задав функцию `salary_group()`, которая присвоит уровень заработной платы в соответствии со значениями в столбцах `salary_from` и `salary_to`.

```python
def salary_group(dataset):
    """Присваиваем уровень заработной платы вакансии в соответствии со значениями в столбцах salary_from и salary_to.
    
    Передаем функции датасет.
    Проверяем входит ли значение в строке в столбцах salary_from и salary_to в указанный диапазон.
    В зависимости от полученного совпадения присваиваем уровень заработной платы.
        
    """
    try:
        if dataset['salary_from'] < 100000 and 100000 <= dataset['salary_to'] < 200000:
            return 'От 100 тысяч до 200 тысяч'
        elif dataset['salary_from'] < 200000 and 200000 <= dataset['salary_to'] < 300000:
            return 'От 200 тысяч до 300 тысяч'
        elif dataset['salary_from'] < 300000 and dataset['salary_to'] >= 300000:
            return 'Больше 300 тысяч'
        elif 0 <= dataset['salary_from'] < 100000 or 0 <= dataset['salary_to'] < 100000:
            return 'Меньше 100 тысяч'
        elif 100000 <= dataset['salary_from'] < 200000 or 100000 <= dataset['salary_to'] < 200000:
            return 'От 100 тысяч до 200 тысяч'
        elif 200000 <= dataset['salary_from'] < 300000 or 200000 <= dataset['salary_to'] < 300000:
            return 'От 200 тысяч до 300 тысяч'
        elif dataset['salary_from'] >= 300000 or dataset['salary_to'] >= 300000:
            return 'Больше 300 тысяч'
        else:
            return 'ЗП не указана'
    except:
        pass
```

Используем заданную функцию - создаём новый столбец `salary_bin_new` со скорректированными уровнями заработной платы и выводим часть датасета на экран для проверки работы функции.

```python
# Используем новую функцию и создаём новый столбец со скорректированными уровнями з/п
vacancies_da_ds['salary_bin_new'] = vacancies_da_ds.apply(salary_group, axis=1)

# Выводим датасет на экран
vacancies_da_ds.head(20)
```

Функция верно присвоила уровни заработной платы. Сравним количество уникальных значений в столбцах `salary_bin` и `salary_bin_new`.

#### В разрезе месяцев публикации

Для расчёта вышеуказанных показателей создадим дополнительный столбец с месяцем публикации вакансии, а также сводную таблицу и присвоим её переменной `vacancies_da_ds_by_month`.

```python
# Создадим дополнительный столбец с месяцем публикации вакансии
vacancies_da_ds['month_of_publishing'] = vacancies_da_ds['published_at'].dt.month

# Создаем сводную таблицу c расчетом показателей по столбцам salary_from и salary_to
vacancies_da_ds_by_month = (
    vacancies_da_ds
    .query('salary_from > 0')
    .pivot_table(values=['salary_from', 'salary_to'], index='month_of_publishing', aggfunc=['mean', 'median', 'count'], 
    margins=True, margins_name='Итого')
    .droplevel(level=1, axis=1)
    .set_axis(['Ср. знач. нижнего порога з/п', 'Ср. знач. верхнего порога з/п',
               'Медиана нижнего порога з/п', 'Медиана верхнего порога з/п',
               'Кол-во вакансий с указанным нижним порогом з/п',
               'Кол-во вакансий с указанным верхним порогом з/п'], 'columns')
    .rename_axis('Месяц публикации вакансии', axis=0)
)

# Выводим созданную сводную таблицу на экран
vacancies_da_ds_by_month
```

## Определение типичного места работы для Аналитика данных и Специалиста по Data Science

Определим типичные места работы Аналитиков данных и Специалистов по Data Science по следующим параметрам:
- ТОП-работодателей по числу вакансий;
- зарплата;
- тип занятости;
- график работы.

Для упрощения выбора необходимых данных и их вывода на экран зададим функцию `barplots_for_grades`, которая будет выводить диаграммы, отображающие информацию о количестве вакансий в зависимости от переданного параметра, в разрезе грейдов.

```python
def barplots_for_grades(dataset, parameter, title):
    """Строим диаграммы, отображающие информацию о количестве вакансий в зависимости от переданного параметра,
    в разрезе грейдов
    
    Передаем функции датасет, необходимый параметр и его название.
    Создаем "таблицу" для диаграмм с двумя строками и четырьмя столбцами, делаем общую подпись.
    Создаем дополнительные датасеты для построения диаграмм путем перебора значений в исходном датасете по типам
    вакансий и в зависимости от грейда представленного в переменной experience_grades, после чего подсчитываем
    количество вакансий по уникальным значениям переданного параметра, реиндексируем столбец и выводим только
    первые 10 позиций.
    Cтроим столбчатые диаграммы по грейдам.
    Выводим на экран готовые диаграммы.
        
    """
    # Создаем "таблицу" для диаграмм с двумя строками и четырьмя столбцами
    fig, ax = plt.subplots(2, 4, figsize = (20, 10))

    # Создаем общую подпись для всех диаграмм
    plt.suptitle(title, fontweight = 'bold')
   
    # Создаем дополнительные датасеты для построения диаграмм путем перебора значений в исходном датасете по типам
    # вакансий и в зависимости от грейда представленного в переменной experience_grades, после чего подсчитываем
    # количество вакансий по уникальным значениям переданного параметра, реиндексируем столбец и выводим только
    # первые 10 позиций
    for i in range(len(experience_grades)):
        da_info = (dataset.loc[(dataset['vacancy_type'] == 'data_analyst') &
                               (dataset['experience'] == f'{experience_grades[i]}'), parameter]
                   .value_counts()).reset_index().head(10)
        ds_info = (dataset.loc[(dataset['vacancy_type'] == 'data_scientist') &
                               (dataset['experience'] == f'{experience_grades[i]}'), parameter]
                   .value_counts()).reset_index().head(10)

        # Строим столбчатую диаграмму для Аналитиков данных
        sns.barplot(x = parameter, y = 'index', data = da_info, ax = ax[0, i], color ='steelblue')
        ax[0, i].set_title(f"Аналитики {experience_grades[i]}")
        ax[0, i].set_xlabel('Количество вакансий, ед.')
        ax[0, i].set_ylabel('')
        
        # Строим столбчатую диаграмму для Специалистов по Data Science
        sns.barplot(x = parameter, y = 'index', data = ds_info, ax = ax[1, i], color ='darkorange')
        ax[1, i].set_title(f"Data Science {experience_grades[i]}")
        ax[1, i].set_xlabel('Количество вакансий, ед.')
        ax[1, i].set_ylabel('')
        
    # Выводим диаграммы на экран
    plt.subplots_adjust(wspace = 1)
    plt.show()
```

Функции задана. Перейдем к анализу вакансий.

Для расчёта помесячной динамики количества вакансий для Аналитиков данных и Специалистов по Data Science создадим новую переменную `vacancies_monthly_dynamics` и запишем в нее столбцы `id`, `experience`, `published_at` и `vacancy_type` из датасета `vacancies_da_ds`, а затем построим линейные диаграммы по аналогии с функцией `barplots_for_grades()`.

```python
# Записываем в переменную vacancies_monthly_dynamics основную информацию по публикации вакансий
vacancies_monthly_dynamics = (
    vacancies_da_ds[['id', 'experience', 'published_at', 'vacancy_type']]
    .sort_values(by='published_at')
)

# Исключаем время из дат публикаций вакансий
vacancies_monthly_dynamics['published_at'] = vacancies_monthly_dynamics['published_at'].dt.date

# Создаем "таблицу" для диаграмм с четырьмя строками и одним столбцом
fig, ax = plt.subplots(4, 1, figsize = (15, 15))

# Создаем общую подпись для всех диаграмм
plt.suptitle('Динамика количества вакансий для Аналитиков данных и Специалистов по Data Science', fontweight = 'bold')

# Создаем дополнительные датасеты для построения диаграмм путем перебора значений в исходном датасете по типам
# вакансий и в зависимости от грейда представленного в переменной experience_grades, после чего подсчитываем
# количество вакансий по уникальным значениям переданного параметра
for i in range(len(experience_grades)):
    da_info = (
        vacancies_monthly_dynamics.loc[(vacancies_monthly_dynamics['vacancy_type'] == 'data_analyst') &
                                       (vacancies_monthly_dynamics['experience'] == f'{experience_grades[i]}'),
                                       'published_at'].value_counts())
    ds_info = (
        vacancies_monthly_dynamics.loc[(vacancies_monthly_dynamics['vacancy_type'] == 'data_scientist') &
                                       (vacancies_monthly_dynamics['experience'] == f'{experience_grades[i]}'),
                                       'published_at'].value_counts())

    # Соединяем датасеты при помощи метода concat и переименовываем столбцы
    da_ds_info = pd.concat([da_info, ds_info], sort=True, axis=1).reset_index().fillna(0)
    da_ds_info.columns = ['Дата','Аналитики данных', 'Специалисты по Data Science']

    # Строим линейную диаграмму для Аналитиков данных и Специалистов по Data Science
    da_ds_info.plot(x = 'Дата', y = ['Аналитики данных', 'Специалисты по Data Science'], style='o-', ax = ax[i], grid=True)
    ax[i].set_title(f"{experience_grades[i]}")
    ax[i].set_xlabel('Дата')
    ax[i].set_ylabel('Количество вакансий, ед.')

plt.subplots_adjust(hspace = 0.4)
```

На основании сформированных линейных диаграмм можно сделать вывод о том, что спрос на специалистов различается по типам вакансий и грейдам:
- среди специалистов без опыта и с опытом работы 1-3 года на протяжении всего рассматриваемого периода (с марта по июнь 2024 г.) чаще всего ищут Аналитиков данных;
- среди специалистов с опытом работы 3-6 лет и более 6 лет чаще всего ищут Специалистов по Data Science;
- в определенные периоды наблюдаются скачки роста спроса на специалистов обоих типов вакансий - в середине апреля, ближе к середине мая и в начале-середине июне, что может быть связано с переходом специалистов из одних компаний в другие и необходимостью закрыт нехватку персонала, а также желанием компаний расширить штат сотрудников для дальнейшего развития.
